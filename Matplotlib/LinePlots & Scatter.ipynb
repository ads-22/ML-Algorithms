# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1)
```

```python
plt.show()
```

```python
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y1)
plt.show()
plt.plot(x,y2)
plt.show()
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("dark_background")
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
plt.style.use("seaborn")
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.plot(x,y1,color='red',label='apple')
plt.plot(x,y2,color='green',label='kiwi')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Price of fruits over Time")
plt.legend()
plt.show()
```

```python
plt.plot(x,y1,color='green',linestyle='dashed',linewidth=1,markersize=24,marker='o')
plt.show()
```

```python
plt.plot(x,y1,color='green',linestyle='dashed',linewidth=1)
plt.show()
```

```python
plt.plot(x,y1,color='green',linestyle='dashed',linewidth=1,marker='o')
plt.show()
```

```python
prices = np.array([1,2,3,4])**2
print(prices)
```

```python
plt.plot(prices,marker='o')
plt.show()
```

# Scatter Plot

```python
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.scatter(x,y1,color='red',label='apple',marker='^')
plt.scatter(x,y2,color='green',label='kiwi')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Price of fruits over Time")
plt.legend()
plt.show()
```

## Adjust the size of plot

```python
plt.figure(figsize=(5,10))
plt.style.use("fivethirtyeight")
plt.scatter(x,y1,color='red',label='apple',marker='^')
plt.scatter(x,y2,color='green',label='kiwi')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Price of fruits over Time")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.style.use("fivethirtyeight")
plt.scatter(x,y1,color='red',label='apple',marker='^')
plt.scatter(x,y2,color='green',label='kiwi')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Price of fruits over Time")
plt.legend()
plt.show()
```
