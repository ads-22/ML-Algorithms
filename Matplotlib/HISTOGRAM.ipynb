```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=np.random.randn(100)
```

```python
sigma=5
u=100
x=X*sigma + u
x=np.round(x)
print(x)
```

# histogram

```python
plt.style.use('seaborn')
```

```python
plt.hist(x)
plt.xlabel("Marsks Range")
plt.ylabel("Frequency Count Of Students")
plt.show()
```

```python
x2=np.round(X*5 + 40)
print(x2)
```

```python
plt.hist(x,alpha=0.5)
plt.hist(x2,alpha=0.8)
plt.xlabel("Marsks Range")
plt.ylabel("Frequency Count Of Students")
plt.show()
```

```python
x2=np.round(X*5 + 90)
print(x2)
```

```python
plt.hist(x,alpha=0.5,label='Physics')
plt.hist(x2,alpha=0.8,label='histogram')
plt.xlabel("Marsks Range")
plt.ylabel("Frequency Count Of Students")
plt.legend()
plt.show()
```

```python
plt.hist(x,alpha=0.5,label='Physics',density=True,orientation='horizontal')
plt.hist(x2,alpha=0.8,label='histogram',density=True,orientation='horizontal')
plt.ylabel("Marsks Range")
plt.xlabel("Frequency Count Of Students")
plt.legend()
plt.show()
```
