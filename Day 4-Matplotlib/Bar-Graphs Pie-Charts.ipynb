# Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar([0,1,2],[10,20,15])
```

```python
plt.bar([0,5,10],[10,20,15])
plt.show()
```

```python
plt.bar([0,1,2],[10,20,15]) # Current Year
plt.bar([0,1,2],[20,10,12]) # Next year
plt.ylim(0,40)
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use("bmh")
x_coord = np.array([0,1,2])*2
plt.bar(x_coord,[10,20,15],width=0.5,label="Current year",tick_label=['gold','silver','platinum']) # Current Year
plt.bar(x_coord+0.5,[20,10,12],width=0.5,label="Current year") # Next year
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coord = np.array([0,1,2])*2
plt.bar(x_coord-0.25,[10,20,15],width=0.5,label="Current year",tick_label=['gold','silver','platinum'],color='green') # Current Year
plt.bar(x_coord+0.25,[20,10,12],width=0.5,label="Current year",color='orange') # Next year
plt.title("Metal Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-5,6)
plt.legend()
plt.show()
```
