```python
import seaborn as sns
```

### Distribution plot
* Allows to see distribution of one variable i.e how spread   that variable is.

### Dist Plot

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
#kde stands for kernal density estimation
```

```python
sns.distplot(tips['total_bill'],kde=False)
#kde stands for kernal density estimation
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
#default is 10 for bins
#this graph looks like normal distribution
# You can see most values are between 10-20 
# Least number of people are in 40-50
```

```python
sns.kdeplot(tips['total_bill'])
#this gives us the line.
```

### Joint Plot
* It means having relation 
* It shows or you can combine 2 distribution plots into one.
* i.e It's between 2 Variables

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

The upper graph shows the distribution for total_bill
while
the lower one for tip....

Hence we are getting both the graphs together.

AS total_bill increases tip also increasing.
## so we can see a linear trend from this graph.
# /
## incresing slope like this and positive !!

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',color='red')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='green')
# Same thing with  a regression line
# See there is a linear trend
```
